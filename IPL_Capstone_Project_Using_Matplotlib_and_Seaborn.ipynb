```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

warnings.filterwarnings('ignore')

df = pd.read_csv('/content/IPL.csv')
df.head()
```

which team won the most matches

```python
df.head()
```

```python
match_wins = df['match_winner'].value_counts()
match_wins
```

```python
# Here all teams like gujarat, banglore like are index and numeric value like 10, 9,8 are values.
sns.barplot(y=match_wins.index,x=match_wins.values,palette='viridis')
plt.title("Most Match Win By Time")
```

Toss Decision Trends

```python
toss_decision = df['toss_decision'].value_counts()
toss_decision
```

```python
sns.countplot(x=df['toss_decision'],palette='rainbow')
plt.title("Toss Decision Trends")
```

Toss Winner vs Match Winner

```python
count = df[df['toss_winner']==df['match_winner']]['match_id'].count()
percentage = (count * 100)/df.shape[0]
percentage.round(2)
```

How do teams win? (Runs vs Wickets)

```python
won_by = df['won_by'].value_counts()
won_by
```

```python
sns.barplot(x=won_by.index, y=won_by.values, palette='rainbow')
```

Key Player Performance

Most "Player of the Matches" Awards

```python
df.head()
```

```python
Player_of_match = df['player_of_the_match'].value_counts().head(5)
Player_of_match
```

```python
sns.barplot(x=Player_of_match.values, y=Player_of_match.index, palette='rainbow')
plt.title("Most Player of the Match Awards")
```

2 Top Score

```python
df.head()
```

```python
# If we want to work with 2 or more than 2 columns than we used groupby() function.
high = df.groupby('top_scorer')['highscore'].sum().sort_values(ascending=False).head(2)
high
```

```python
high.plot(kind='barh', color='red')
plt.title("Top Scorer")
```

10 Best Bowling Figures

```python
df.head()
```

```python
df['highest_wickets'] = df['best_bowling_figure'].apply(lambda x: x.split('--')[0])
df['highest_wickets'] = df['highest_wickets'].astype(int)
top_bowlers = df.groupby('best_bowling')['highest_wickets'].sum().sort_values(ascending=False).head(10)
```

```python
top_bowlers.plot(kind='barh',color='red')
```
