```python
#  Matplotlib
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Matplotlib Basics
x=np.linspace(0,5,11)
x
y=x**2
y
```

```python
plt.plot(x,y)
plt.title("Hello")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
```

```python
# Subplots in Matplotlib: used for creating multiple plots
```

```python
plt.subplot(2,2,1)
plt.plot(x**2,y**4)
plt.subplot(2,2,4)
# in 4th plot we want to create something
plt.plot(x,y)
plt.subplot(2,2,3)
plt.plot(x**2,y**2)
plt.subplot(2,2,2)
plt.plot(y,x)
```

```python
#  Object Oriented Way
```

```python
fig = plt.figure(figsize=(6,7),dpi=100)
# figsize is a tuple of the width and height of the figure in inches dpi is the dots-per-inch(pixel per inch).
# here they will take 4 values left,bottom,width,height and the value is always beetween 0 and 1
axis1 = fig.add_axes([0.1,0.1,0.8,0.8])
axis1.plot(x,y)
axis2 = fig.add_axes([0.25,0.33,0.2,0.2])
axis2.plot(x,y)

# Here axis1 will be overlap with axis2
```

```python
#  Types of Plots
```

```python
# Scatter Plots
plt.scatter(x,y)
```

```python
# Histogram
from random import sample
data = sample(range(1,1000),100)
plt.hist(data)
```

```python
# Boxplot
plt.boxplot(x)
```

```python
# Saving Plots
fig = plt.figure()
ax = fig.add_axes([0.1,0.1,0.8,0.8])
ax.plot(x,y)
# save the image in the png or any format.
plt.savefig("basicplot.jpg")
```
