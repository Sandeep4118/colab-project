```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Matplotlib Basics
x=np.linspace(0,5,11)
x
y=x**2
y
```

```python
plt.plot(x,y)
plt.title("Hello")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
```

```python
# Subplots in Matplotlib: used for creating multiple plots
```

```python
plt.subplot(2,2,1)
plt.plot(x**2,y**4)
plt.subplot(2,2,4)
# in 4th plot we want to create something
plt.plot(x,y)
plt.subplot(2,2,3)
plt.plot(x**2,y**2)
plt.subplot(2,2,2)
plt.plot(y,x)
```
