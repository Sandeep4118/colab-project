```python
import numpy as np
from scipy.stats import norm
```

```python
sample = [172,174,168,171,173,175,170,169,172,172,174,168,171,173,175,170,169,172,172,174,168,171,173,175,170,169,172,174]
population_mean = 170
population_std = 3
sample_mean = np.mean(sample)
n = len(sample)
```

```python
z_score = (sample_mean - population_mean)/(population_std/np.sqrt(n))
z_score
```

```python
# finding of p-value
p_value = 2 * (1 - norm.cdf(abs(z_score)))
p_value
```

```python
alpha = 0.05
# if p_value < alpha then we reject the hypothesis but if the p_value > alpha the we accept the hypothesis.
if p_value < alpha:
  print("I will reject the null hypothesis")
else:
  print("I will accept the null hypothesis")
```
