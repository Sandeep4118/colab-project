```python
# GroupBy
```

```python
data = {
    'Category': ['A', 'B', 'A', 'B', 'A', 'B', 'A', 'B'],
    'Store': ['S1', 'S1', 'S2', 'S2', 'S1', 'S2', 'S2', 'S1'],
    'Sales': [100, 200, 150, 250, 120, 180, 200, 300],
    'Quantity': [10, 15, 12, 18, 8, 20, 15, 25],
    'Date': pd.date_range('2023-01-01', periods=8)
}

df = pd.DataFrame(data)
```

```python
df
```

```python
# Group by Category and calculate the sum of Sales.
cat = df.groupby('Category')['Sales'].sum()
cat
# Group by Creates an object which we can't see it.
```

```python
# Group by store and calculate the sum of Sales.
cat = df.groupby('Store')['Sales'].sum()
cat
```

```python
# Group by multiple columns
# Group by Category and Store
# we do groupby for multiple columns for basically categorical data not a continious data.
# There are two type of data qulatative and quantative, here qulatative means categorial data(like male and female, class A and class B, S1 and S2 etc) and quantative means numerical data like quantity,sales etc.
cat = df.groupby(['Category','Store'])['Sales'].sum()
cat
```

```python
# Aggregation : finding the value from the multiple things.
# like mean, median,sd, variance etc
df['Sales'].mean()
```

```python
# we have aggregation function(agg) also where we can pass all the aggreatation type in one way.
df['Quantity'].agg(['mean','min','max','median','count','std'])
```

```python
# loading the data
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('/content/anime.csv')
df
```

```python
# to get only top 5 rows
df.head()
```

```python
# Make a new column for episode count, you will see that brackets (), where the episode is written.
# Feature Extraction
df.loc[2]['Title']
```

```python
def extract_episode(txt):
  check = False
  data = ""
  for i in txt:
    if i == ')':
      # data +=i
      break
    if i == '(':
      check = True
      continue
    if check==True:
      data += i
  return data
```

```python
df['episode_count'] = df['Title'].apply(extract_episode)
```

```python
df
```

```python
df['episode_count'] = df['episode_count'].str.replace("eps","")
```

```python
df
```

```python
# Convert episode_count into integer because it is actually a string.
df['episode_count']=df['episode_count'].astype(int)
```

```python
df
```

```python
# Make a new column for time stamp.
df.loc[1]['Title']
```

```python
def extract_timestamp(txt):
  check = False
  data = ""
  for i in range(len(txt)):
    if txt[i] == ')':
      for j in range(i+1,i+20):
        data += txt[j]
        check = True
    if check == True:
      break
  return data
```

```python
df['episode_timestamp'] = df['Title'].apply(extract_timestamp)
```

```python
df
```

```python
from dateutil.relativedelta import relativedelta
from datetime import datetime

def calculate_total_months(period):
    try:
        start_str, end_str = period.split(' - ')
        start_date = datetime.strptime(start_str, '%b %Y')
        end_date = datetime.strptime(end_str, '%b %Y')
        r = relativedelta(end_date, start_date)
        return r.years * 12 + r.months + 1  # +1 to include the starting month
    except:
        return None

df['Months'] = df['episode_timestamp'].apply(calculate_total_months)
```

```python
df
```

```python
# which anime has the highest score
df[df['Score'] == df['Score'].max()]
```

```python
# give me top 5 highest scoring anime
# here we use value_counts() it gives the value in desiding order.
# head() gives the value of top five rows
df['Score'].value_counts().head()
```

```python
# which anime has the highest episode count
df[df['episode_count'] == df['episode_count'].max()]
```

```python
# animes with top 5 episode count
df['Title'].value_counts().head()
```

```python
# which is the longest running anime
df[df['Months'] == df['Months'].max()]
```
