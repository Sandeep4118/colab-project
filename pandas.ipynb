```python
# Missing Data
# FInding Missing Data
import numpy as np
import pandas as pd
```

```python
data = {
    'A':[1, 2, np.nan, 4, 5],
    'B':[1, 2, 3, 4, 5],
    'C' :[1, 2, 3, np.nan,np.nan],
    'D' :[1,np.nan, np.nan,np.nan,5]
}
df = pd.DataFrame(data)
```

```python
df
```

```python
df.isna()
```

```python
# NAN - is a null value that's why it's show true.
df.isna().sum()
```

```python
# any missing value in each coloumn.
df.isna().any()
```

```python
# Remvoing Missing Data works only on rows
```

```python
df
```

```python
df.dropna()
```

```python
df
```

```python
# threshold mean there is given 3 means it must be 3 (not null value) in the rows.
df.dropna(thresh=3)
```

```python
# Filing the missing Data
```

```python
# Instead of NAN fill the another value
df
```

```python
df.fillna(0)
# it doesn't apply to the original df because if you want to change original df you must use inplace=true so that it will change the NAN to 0 permenentatly.
```

```python
df
```

```python
values = {'A':0,'B':100,'C':300,'D':400}
# coloumn wise
df.fillna(value=values)
```

```python
df
```

```python
# if you want to put that value of NAN that is the mean of that table.
df.fillna(df.mean())
```
