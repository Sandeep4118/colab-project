```python
# Distribution plot for contigous variable like numerical value.
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = sns.load_dataset('tips')
df
```

```python
df['size'].unique()
```

```python
# histplot type of Distribution plot.
# histogram which is used for showcasing of data.
# if the bins=1 then all the value goes into only one bin.
# and if the bins=3 then all the value goes into 3 bins according their value.
sns.histplot(df['total_bill'],bins=20,kde=True)
```

```python
plt.subplot(1,2,1)
sns.histplot(df['total_bill'],bins=20,kde=True)
plt.subplot(1,2,2)
sns.histplot(df['tip'],bins=20,kde=True)
```

```python
# jointplot
```

```python
sns.jointplot(data=df,x='total_bill',y='tip',kind='scatter')
```

```python
sns.jointplot(data=df,x='total_bill',y='tip',kind='hist')
```

```python
sns.jointplot(data=df,x='total_bill',y='tip',kind='kde')
```

```python
sns.jointplot(data=df,x='total_bill',y='tip',kind='hex')
```

```python
sns.jointplot(data=df,x='total_bill',y='tip',kind='reg')
```

```python
# Pair Plot
```

```python
sns.pairplot(df,hue='size',palette='rainbow')
```

```python
# rugplot
```

```python
sns.rugplot(df)
```
