```python
#  Categorical Data Plots for categorical value.
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = sns.load_dataset('tips')
```

```python
df
```

```python
# Types of Categorical Data Plots
# boxplot
# vinlinplot
# stripplot
# swarmplot
# barplot
# countplot
```

```python
# countplot : if we want to find the no. of males and female and which one smoke more. then we use countplot.
```

```python
sns.countplot(x = df['sex'],hue = df['smoker'])
```

```python
# barplot: Here on y-axis all numerical values show's and on x-axis all categrocial value's show
```

```python
sns.barplot(x=df['sex'],y=df['tip'])
```

```python
# You can change the estimator object to your own function, that converts a vector to a scalar
# Mean (default) : np.mean
# Meadian: np.median
# Sum : np.sum
# Count : len(to count the number of observations)
# Standard Deviation : np.std
# Minimum : np.min
# Maximum : np.max
```

```python
import numpy as np
sns.barplot(x=df['sex'],y=df['tip'],estimator=np.sum)
# It gives the total estimated tip of male and female seperately.
```

```python
sns.barplot(x=df['sex'],y=df['total_bill'],estimator=np.sum)
```

```python
# boxplot : Boxplot is a graph that shows the distribution of data and helps identify outliers.
# boxplot are builts using 5 number summary in Statics.
```

```python
sns.boxplot(x=df['tip'],y=df['day'],data=df,palette='rainbow')
```

```python
# violinplot : Violin plot is a graph that shows the distribution of data along with its shape and density.
```

```python
# Boxplot/Violinplot = numerical data ko different categories me compare karne ke liye.
```

```python
sns.violinplot(x=df['day'], y=df['total_bill'],data=df)
```

```python
# Stripplot : Strip plot shows individual data points as dots for each category.
# Striplot : dots thode overlap ho sakte hain.
# Swarmplot : dots perfectly arranged (no overlap).
# Swarmplot : Swarm plot also shows individual points but arranged neatly without overlap.
# Stripplot = scattered dots
# Swarmplot = well-organized dots

# Categorical vs Numerical data ke individual points dekhne ke liye.
```

```python
sns.stripplot(x=df['day'], y=df['total_bill'],data=df)
```

```python
sns.swarmplot(x=df['day'], y=df['total_bill'],data=df)
```

```python
# combining violinplot and swarmplot together.
sns.violinplot(x='tip',y='day',data=df,palette='rainbow')
sns.swarmplot(x='tip',y='day',data=df,palette='bright')
```
